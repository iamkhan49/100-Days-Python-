**DAY 16** Match Case

```python
#Used to check one value against multiple cases (Cleaner alternative to many if else elif)

day = int(input("Enter the day number: "))

match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
  case _:                        #case_ : works like else
    print("Invalid Day")
```

```python
day2 = int(input("Enter the day number: "))

match day2:
  case 1 | 2 | 3 | 4 | 5:
    print("Weekday")
  case 6 | 7:
    print("Weekend")
```

```python
#we can do everthing with if-else-elif that match case can do. Python added match-case because it become cleaner when ther are many possible values
#Main use of match case is to make code cleaner
#With if else elif

day3 = int(input("Enter the day number: "))

if day3 == 1:
  print("Monday")
elif day3 == 2:
  print("Tuesday")
elif day3 == 3:
  print("Wednesday")
else:
  print("Invalid Day")
```

```python
#Math case with if

x = int(input("Enter the value of x: "))

match x:

  case 0:
    print("x is Zero")

  case x if x == 4:
    print("x is 4")

  case _ if x != 90:
    print(x, "is not 90")

  case _ if x != 80:
    print(x, "is not 80")

  case _:
    print(x)
```
